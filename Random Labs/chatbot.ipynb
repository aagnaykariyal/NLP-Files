```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import sent_tokenize
import re
```

```python
file = open("file.txt", "r")
documents = sent_tokenize(file.read())
len(documents)
```

```python
# Data Cleaning
def clean_text(txt):
    sent = re.sub(r"\[.+\]", '', txt.lower())
    sent = sent.replace("\n", " ")
    sent = sent.replace("  ", "")
    return sent
```

```python
# Cleaning the corpus
cleaned_doc = []
for docs in documents:
    cleaned_doc.append(clean_text(docs))
```

```python
cleaned_doc
```

```python
def get_output(query):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(cleaned_doc)
    docs = []
    query_vector = vectorizer.transform([query])
    
    cs = cosine_similarity(query_vector, tfidf_matrix)
    result = cs[0].argsort()[-10][::-1]
    result = result.tolist()
```
